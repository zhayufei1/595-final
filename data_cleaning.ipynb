```
str1 = file_contents
RESULTS = str1.split()
print(RESULTS)
```

```
import csv
abc = RESULTS
new=[(abc[0],)]+[(''.join(i),) for i in zip(abc[1::2],abc[2::2])]

import csv
with open('test.csv', 'w', newline='') as fp:
    a = csv.writer(fp, delimiter=',')
    a.writerows(new)

    
```
